### What country has the highest percentage of people that earn >50K and what is that percentage?
##### the answer is 
|country | >50k |
|--------|------|
|Iran    |41.9 %|

```python
country_per_salary  = df.groupby('native-country')['salary'].value_counts(normalize=True).unstack()
country_percentage_rich = (country_per_salary.get('>50K',0)*100).fillna(0)
highest_earning_country = country_percentage_rich.idxmax()
highest_earning_country_percentage = round(country_percentage_rich.max(),1)
highest_earning_country_percentage
print(highest_earning_country, highest_earning_country_percentage)
```

```python
countr = df.groupby('native-country')['salary'].value_counts(normalize=True).unstack()
# countr
per = countr['>50K']
salary = pd.DataFrame(per)
percentage = round((salary.sort_values(by='>50K', ascending=False).head(1))*100,1)
percentage
```
